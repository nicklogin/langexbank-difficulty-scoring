```python
import pandas as pd
df1 = pd.read_excel("realec_errors_sample.xlsx", engine='openpyxl')[['id', 'Sentence']]
df2 = pd.read_excel("realec_sample_concat.xlsx", engine='openpyxl')[['Sentence','Right answer',
                                                                   'Error type','Corrected',
                                                                   'Folder','Filename']]
```

```python
df = df1.merge(df2, on="Sentence")
```

```python
df
```

```python
df = df.drop_duplicates(subset=['Sentence'])
```

```python
df
```

```python
prefix_len = len('downloaded_2021_08_05_11_15_07094784/data')
df['path_in_REALEC'] = df.apply(lambda x: "https://realec.org/index.xhtml#"+x['Folder'][prefix_len:]+'/'+x['Filename']+'\n',
                                axis=1)
```

```python
df
```

```python
df = df.drop(['Folder','Filename'], axis=1)
```

```python
df.to_excel("realec_sample_concat_final.xlsx", engine="openpyxl")
```
